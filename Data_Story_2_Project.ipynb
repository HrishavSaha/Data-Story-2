```python
import pandas as pd
import statistics as st
import random
import plotly.figure_factory as ff
import plotly.graph_objects as go
import numpy as np
```

```python
df = pd.read_csv('fmsave.csv')
savings = df['quant_saved']
df.head()
```

```python
#Calculating IQR
q1 = savings.quantile(0.25)
q3 = savings.quantile(0.75)
iqr = q3 - q1
print(f'IQR: {iqr}')
```

```python
#Removing outliers from data
l_whisker = q1 - (1.5 * iqr)
u_whisker = q3 + (1.5 * iqr)

updated_df = df[df['quant_saved'] < u_whisker]
updated_savings = updated_df['quant_saved']
updated_df.head()
```

```python
#Basic statistical calculations for updated dataframe
mean = st.mean(updated_savings)
median = st.median(updated_savings)
mode = st.mode(updated_savings)
stdev = st.stdev(updated_savings)

print(f'mean: {mean}')
print(f'median: {median}')
print(f'mode: {mode}')
print(f'stdev: {stdev}')
```

```python
#Making a sample and plotting it
iterations = 100
temp = updated_savings.tolist()
sample = []

for i in range(iterations):
  rand = random.randint(0, len(temp) - 1)
  sample.append(temp[rand])
  temp.pop(rand)

sample_mean = st.mean(sample)

fig = ff.create_distplot([sample], ["Savings (Sample)"], show_hist=False)
fig.add_trace(go.Scatter(x=[sample_mean, sample_mean], y=[0, 0.01], mode="lines", name="MEAN"))
fig.show()
```

```python
#Stdev of sample
stdev_sample = st.stdev(sample)
print('Sample stdev:', stdev_sample)
```

```python
#Mean of population and sample
print('population mean:', mean)
print('sample mean', sample_mean)
```
